# gen_exec.ipynb
    - batch script generator for PO4 maps
    - updated in Feb 2026

```python
import numpy as np
import os
import time
```

```python
with open('exec.sh', 'w') as text_file:
    cnt=0
    for n3 in np.arange(1,6,1):
        for n2 in [1]:
            for n1 in [2]:
                for n5 in np.arange(2,4,1):
                    text_file.write('#--------------------------------------------------------\n')
                    cnt=cnt+1
                    vnow=f'{n1}.{n2}.{n3}.1.{n5}.4'
                    com =f'echo \'{vnow}\' > data_{cnt}.txt\n'
                    text_file.write(com)
                    com = f'cp ../src/run.pbs run{cnt}.pbs\n'
                    text_file.write(com)
                    com = f'sed -i \'s/po4_train.py/po4_train{cnt}.py/g\' run{cnt}.pbs\n'
                    text_file.write(com)
                    com = f'sed -i \'s/po4_eval.py/po4_eval{cnt}.py/g\' run{cnt}.pbs\n'
                    text_file.write(com)
                    com = f'sed -i \'s/po4_project.py/po4_project{cnt}.py/g\' run{cnt}.pbs\n'
                    text_file.write(com)
                    com = f'sed -i \'s/map_po4_XXX/map_po4_{cnt}/g\' run{cnt}.pbs\n'
                    text_file.write(com)
                    com = f'cp ../src/po4_train.py po4_train{cnt}.py\n'
                    text_file.write(com)
                    com = f'cp ../src/po4_eval.py po4_eval{cnt}.py\n'
                    text_file.write(com)
                    com = f'cp ../src/po4_project.py po4_project{cnt}.py\n'
                    text_file.write(com)
                    com = f'sed -i \'s/data_XXX/data_{cnt}/g\' po4_train{cnt}.py\n'
                    text_file.write(com)
                    com = f'sed -i \'s/data_XXX/data_{cnt}/g\' po4_eval{cnt}.py\n'
                    text_file.write(com)
                    com = f'sed -i \'s/data_XXX/data_{cnt}/g\' po4_project{cnt}.py\n'
                    text_file.write(com)
                    com = f'qsub run{cnt}.pbs\n'
                    text_file.write(com)
                    text_file.write('sleep 3\n')
```
